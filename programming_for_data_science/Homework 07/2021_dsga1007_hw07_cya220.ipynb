**1.10 (15 points)**

Use the ratings `rating_final.csv`. 'Group by' the data by `placeID` and obtain `mean` and `std` of each of the ratings (`rating`, `food_rating`, and `service_rating`) for each restaurant (i.e. for each `placeID`).

Print the head of this frame so to show the results. 

```python
df_1 = rating_final.groupby(['placeID'])[['rating','food_rating','service_rating']].mean().add_suffix('_mean')
df_2 = rating_final.groupby(['placeID'])[['rating','food_rating','service_rating']].std().add_suffix('_std')
df = pd.concat((df_1, df_2), axis=1)
df.head()
```

**1.11 (10 points)**


Rank the restaurants IDs (i.e. 'placeID') by their `mean ratings` increasingly.

Print the restaurants IDs (i.e. 'placeID') and `mean ratings` with the lowest `mean ratings` (hint: lowest appears at the top).

Print the restaurants IDs (i.e. 'placeID') and `mean ratings` with the highest `mean ratings` (hint: highest appears at the bottom).



```python
mean_ratings = df[['rating_mean']].sort_values(by=['rating_mean'])
print("The lowest mean rating:")
print(mean_ratings.head(1))
print("The highest mean rating:")
print(mean_ratings.tail(1))
```

**1.12 (10 points)**

Make and print a table with columns (`place`, `mean`, `name`), 
which are the restaurant's ID, mean ratings, and name, 
of the 10 restaurants with the lowest `mean ratings`.


```python
restaurant_names = geoplaces2[['placeID','name']]
ratings_names = mean_ratings.head(10).merge(restaurant_names, how='left', on='placeID')
ratings_names = ratings_names.rename(columns={"placeID": "place", "rating_mean": "mean"})
ratings_names
```

**1.13 (10 points)**

Make and print a table with columns (`place`, `mean`, `name`), 
which are the restaurant's ID, mean ratings, and name, 
of the 10 restaurants with the highest `mean ratings`.


```python
restaurant_names = geoplaces2[['placeID','name']]
ratings_names = mean_ratings.tail(10).merge(restaurant_names, how='left', on='placeID')
ratings_names = ratings_names.rename(columns={"placeID": "place", "rating_mean": "mean"})
ratings_names
```

**1.14 (10 points)**

```python
restaurant_info = geoplaces2[['placeID','name','city','state']]
ratings_names = mean_ratings.merge(restaurant_info, how='left', on='placeID')
#ratings_names
print("The restaurant with the lowest mean rating is",ratings_names.name[0],
      "which is located in",ratings_names.city[0]+",",ratings_names.state[0])
print("The restaurant with the highest mean rating is",ratings_names.name[len(ratings_names)-1],
      "which is located in",ratings_names.city[len(ratings_names)-1]+",",ratings_names.state[len(ratings_names)-1])
#print("The highest mean rating:")
#print(ratings_names.tail(1))
```
